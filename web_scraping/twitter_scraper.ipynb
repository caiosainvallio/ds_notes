# Importar tweets do bonobo

```
from twitter_scraper import get_tweets
tuites = []
for t in get_tweets('jairbolsonaro'):
  d = {'text':t['text'], 'likes':t['likes'], 'retweets':t['retweets']}
  tuites.append(d)
```

# Visualizar os tuítes com maior número curtidas

```
def criterio(tuite):
  return tuite['likes']

tuites_ordenados = sorted(tuites, key=criterio, reverse=True)
for t in tuites_ordenados:
  print(t['likes'], t['text'], '\n', '-'*20, '\n')
```

# Visualizar os tuítes com maior número de retweets

```
def criterio(tuite):
  return tuite['retweets']

tuites_ordenados = sorted(tuites, key=criterio, reverse=True)
for t in tuites_ordenados:
  print(t['retweets'], t['text'], '\n', '-'*20, '\n')
```

```
from twitter_scraper import get_tweets
import markovify

textos = []
for t in get_tweets('jairbolsonaro'):
  textos.append(t['text'])

text_model = markovify.Text(textos)
```

```
print(text_model.make_short_sentence(140))
```
